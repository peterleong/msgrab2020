```
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Input
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
import numpy as np 
```

```
safety_df = pd.read_csv('Safety_DataSet_Aggregated.csv')
safety_df
```

```
safety_df['label'].value_counts()
```

```
balanced_label0_df = safety_df[safety_df['label'] == 0].sample(n=len(safety_df[safety_df['label'] == 1]))
balanced_label0_df
```

```
balanced_label1_df = safety_df[safety_df['label']==1]
balanced_label1_df
```

```
final_df = pd.concat([balanced_label0_df,balanced_label1_df])
final_df = final_df.reset_index(drop=True)
final_df
```

```
final_df['accelerometer_data'] = np.sqrt( (final_df['acceleration_x']) **2 + (final_df['acceleration_y']) **2 + (final_df['acceleration_z']) **2)
final_df 
```

```
features = list(final_df.columns.values)
features.remove('label')
features.remove('bookingID')
features.remove('acceleration_x')
features.remove('acceleration_y')
features.remove('acceleration_z')
```

```
y = final_df['label']
y
```

```
from sklearn.decomposition import PCA
pca_gyro = PCA().fit(final_df.loc[:, ['gyro_x', 'gyro_y', 'gyro_z']])
pca_gyro.explained_variance_ratio_
```

```
features.remove('gyro_z')
X = final_df[features]
X
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```
# Standard Scaler
ss = StandardScaler()
X_train_standardised = ss.fit_transform(X_train)
X_test_standardised = ss.transform(X_test)
```

```
X_train_standardised
```

```
X_test_standardised
```

```
model = Sequential()
model.add(Dense(100, input_dim=X_train_standardised.shape[1], activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```
model.summary()
```

```
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train_standardised, y_train, epochs=20, batch_size=10)
test_loss, test_acc = model.evaluate(X_test, y_test)
```

```
# Get actual predicted value 
print("CNN Error: %.2f%%" % (100-test_acc*100))
```
